```python
# 1.On a single plot (use matplotlib.pytplot.subplot), draw two histograms: for female 
#weights (top subfigure). and for male weights (bottom subfigure).

# Determine the x-axis limits
min_weight = min(female_df['Weights'].min(), male_df['Weights'].min())
max_weight = max(female_df['Weights'].max(), male_df['Weights'].max())

# Create the subplots
plt.figure(figsize=(10, 6))

# Top subplot for female weights
plt.subplot(2, 1, 1)
plt.hist(female_df['Weights'], bins=20, color='black')
plt.title('Female Weights')
plt.xlim(min_weight, max_weight)

# Bottom subplot for male weights
plt.subplot(2, 1, 2)
plt.hist(male_df['Weights'], bins=20)
plt.title('Male Weights')
plt.xlim(min_weight, max_weight)

plt.tight_layout()
plt.show()
```

**In the above both histogram are seem to be right-skewed distribution(Positive skew). But male weights are little bit showing Normal Distribution**

**In the histogram of female, weights are getting high  values between 45 and 90.**

```python
# 2.Call matplotlib.pyplot.boxplot to draw a box-and-whisker plot, with two boxes side by side, giving the male and female weights 
#so that they can be compared to each other. 

# Extract the weight columns
female_weights = female_df['Weights']
male_weights = male_df['Weights']

# Create the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot([female_weights, male_weights], labels=['Female', 'Male'])
plt.title('Comparison of Female and Male Weights')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# 3.Compute the basic numerical aggregates of the male and female weights (measures of location, dispersion, and shape). 

# Calculate basic numerical aggregates for female weights
female_mean = female_weights.mean()
female_median = female_weights.median()
female_std = female_weights.std()
female_variance = female_weights.var()
female_skew = female_weights.skew()

print("Female Weights:")
print(f"Mean: {female_mean:.2f}")
print(f"Median: {female_median:.2f}")
print(f"Standard Deviation: {female_std:.2f}")
print(f"Variance: {female_variance:.2f}")
print(f"Skewness: {female_skew:.2f}")

# Calculate basic numerical aggregates for male weights
male_mean = male_weights.mean()
male_median = male_weights.median()
male_std = male_weights.std()
male_variance = male_weights.var()
male_skew = male_weights.skew()

print("\nMale Weights:")
print(f"Mean: {male_mean:.2f}")
print(f"Median: {male_median:.2f}")
print(f"Standard Deviation: {male_std:.2f}")
print(f"Variance: {male_variance:.2f}")
print(f"Skewness: {male_skew:.2f}")
```

#### Comparison of distributions

**The male weight distribution has a slightly higher mean and standard deviation compared to the female weight distribution.**

**Both distributions appear to be slightly right-skewed, with positive skewness values.**

**The kurtosis values indicate that both distributions have slightly heavier tails than a normal distribution.**

```python
# 4.To the female matrix, add the eight column which gives the body mass indices of all the female participants. 

#Calculate BMI
female_df['BMI'] = female_df['Weights'] / ((female_df['Standard_height'] / 100) ** 2)
female_df.head()
```

```python
# 5.Create a new matrix zfemale being a version of the female dataset with all its columns standardised (by computing the z-scores of each column).

# Calculate the mean and standard deviation of each column
female_mean = female_df.mean()
female_std = female_df.std()

# Standardize the data by subtracting the mean and dividing by the standard deviation
zfemale = (female_df - female_mean) / female_std
print(zfemale)
```

```python
# 6.Draw a scatterplot matrix (pairplot) for the standardised versions of height, weight, waist circumference, hip circumference, and BMI of the female 
#Compute Pearsons and Spearmans correlation coefficients for all pairs of variables. 

# Select the columns for the scatterplot matrix
columns_for_pairplot = ['Standard_height', 'Weights', 'Waist_circum', 'Hip_circum', 'BMI']
subset_zfemale = zfemale[columns_for_pairplot]

# Create the scatterplot matrix
sns.pairplot(subset_zfemale)
plt.show()
```

```python
#Compute Pearsons and Spearmans correlation coefficients for all pairs of variables.

# Calculate Pearson's correlation
pearson_corr = subset_zfemale.corr()
print("Pearson's correlation coefficients:\n", pearson_corr)

# Calculate Spearman's correlation
spearman_corr = subset_zfemale.corr(method='spearman')
print("\nSpearman's correlation coefficients:\n", spearman_corr)
```

#### Interpretation

**The scatterplot matrix shows the relationships between the standardized variables.**

**Pearson's correlation measures linear relationships, while Spearman's correlation measures monotonic relationships.**

**Strong positive correlations are observed between weight and BMI, height and BMI, and waist circumference and hip circumference.**

**These correlations suggest that as weight, height, waist circumference, or hip circumference increase, BMI also tends to increase.**

**The correlation coefficients provide a numerical measure of the strength and direction of these relationships.**

#### 9. Advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio:

#### - **BMI:**

**Advantages:**
   - **Easy to calculate and widely used.** 
   - **Provides a general indication of body fatness.** 
   - **Useful for population-level studies.**

**Disadvantages:**
   - **Doesn't account for body composition (muscle vs. fat).**
   - **Not suitable for athletes or individuals with varying body types.**
   - **Doesn't consider regional fat distribution.**

#### - **Waist-to-height ratio:**

**Advantages:**
   - **Simple to calculate.**
   - **Better indicator of visceral fat than BMI.**
   - **Useful for predicting health risks.**

**Disadvantages:** 
   - **Less widely used than BMI.**
   - **May not be as accurate for people with varying body shapes.**

#### - **Waist-to-hip ratio:**

**Advantages:** 
   - **Good indicator of overall body fat distribution.**
   - **Helps assess risk of chronic diseases like heart disease and diabetes.**
 
**Disadvantages:** 
   - **More complex to measure than BMI.**
   - **May not be as accurate for people with unusual body shapes.**
